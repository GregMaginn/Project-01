```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Top Spotify songs from 2010-2019 - global, by year
# source: https://www.kaggle.com/leonardopena/top-spotify-songs-from-20102019-by-year?select=top10s.csv

df = pd.read_csv('resources_hl/top10s.csv', encoding = "ISO-8859-1")
```

```python
df.rename(columns=({'Unnamed: 0': ''})).set_index('')
```

```python
df2 = df[['title', 'artist', 'top genre', 'year']]
df2
```

## What is the most popular genre?

```python
# Top 7 genres across all years
topgenres = df2['top genre'].value_counts()
topgenres[:7]
```

## Dance pop is almost 5.5x more popular than the next genre below it.

```python
colors = ["red", "orange", "lightcoral", "lightskyblue", 'green', 'yellow', 'cyan']
explode = (0.05, 0 ,0 ,0,0,0,0)
fig,ax = plt.subplots(1,figsize=(8,8))
ax.pie(topgenres[:7], labels=topgenres[:7].index, startangle=70, autopct='%.1f%%', explode=explode, shadow=True,
      colors=colors);
ax.set_title('Spotify Global Top 7 Genres from 2010-2019')
plt.tight_layout()
plt.savefig('resources_hl/genres_pie.png')
```

## Which artist appeared on Top Songs the most?

```python
df2_artist = df2.artist.value_counts()
df2_artist[:10]
```

```python
plt.style.use('ggplot') # style gives grid and light gray background
plt.figure(figsize=(9, 5))
bars = plt.bar(df2_artist.index[:10], df2_artist[:10], fc='pink', ec='black', width=.5)
plt.xticks(rotation=90)
plt.ylim(0,20)
plt.title('Number of Songs by Artist on Spotify Top Songs 2010-2019')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
# add values at top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x()+.1, yval + 0.9, yval)

plt.tight_layout()

plt.savefig('resources_hl/popular_artists.png')
```

### Justin Bieber is the only artist with 9 songs appearing in Top Songs in any given year.

```python
artists = ['Katy Perry', 'Justin Bieber', 'Ed Sheeran', 'The Chainsmokers']
filterartist = pd.concat([df2[df2.artist ==  artist].year.value_counts() for artist in artists], axis=1)
filterartist.columns = artists
filterartist.fillna(0, inplace=True)
ax = filterartist.plot()
plt.title('Number of Songs by Artist on Top Songs 2010-2019')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
ax.set_ylim(0,10)
plt.setp(ax, xticks=filterartist.index);
plt.tight_layout()
plt.savefig('resources_hl/artiststhroughyears.png')
```
