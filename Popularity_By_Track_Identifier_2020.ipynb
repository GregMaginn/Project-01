```python
# Import Dependencies

import pandas as pd
```

```python
 # Name of the CSV file
file = "resources_ms/data.csv"
```

```python
# Read the file

df = pd.read_csv(file)
```

```python
# Print df

df.head()
```

```python
# df only including 2020

year = df["year"] == 2020

df_2020 = df.loc[year]

# Reset index

df_2020_final = df_2020.reset_index()

# print final df

df_2020_final
```

```python
# sort by descending popularity

df_2020_final_sorted_popularity = df_2020_final.sort_values(by=["popularity"], axis=0, ascending=False)

df_2020_final_sorted_popularity
```

```python
# Create a dataset showing only 70+ popularity
```

```python
seventy_plus = df_2020_final_sorted_popularity["popularity"] >= 70

seventy_plus_df = df_2020_final_sorted_popularity.loc[seventy_plus]

# Reset index

seventy_plus_df = seventy_plus_df.reset_index()

# print final df

seventy_plus_df
```

```python
# drop strings


seventy_plus_df = seventy_plus_df.drop(["artists", "id", "name", "release_date"], axis=1)

seventy_plus_df
```

```python
# Find average of identifiers of 2020

seventy_plus_df.mean(axis=0)
```

```python
# Run independent t-test

from scipy.stats import ttest_ind
```

```python
# T Test for acousticness

acousticness_popular = seventy_plus_df["acousticness"]


acousticness_all = df_2020_final_sorted_popularity["acousticness"]


acousticness_popular_p_value = ttest_ind(acousticness_popular, acousticness_all)

# Print rounded P Value

acousticness_popular_p_value = round(acousticness_popular_p_value[1], 3)

acousticness_popular_p_value = print(f' P Value: {acousticness_popular_p_value}')

acousticness_popular_p_value
```
