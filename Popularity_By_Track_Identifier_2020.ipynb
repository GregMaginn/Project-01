```python
# Import Dependencies

import pandas as pd
```

```python
 # Name of the CSV file
file = "resources_ms/data.csv"
```

```python
# Read the file

df = pd.read_csv(file)
```

```python
# df only including 2020

year = df["year"] == 2020

df_2020 = df.loc[year]

# Reset index

df_2020_final = df_2020.reset_index()

# print final df

df_2020_final

x = df_2020_final["popularity"].unique()
x.sort()
x

```

```python
# Run independent t-test

from scipy.stats import ttest_ind
```

```python
# T Test for acousticness

acousticness_popular = seventy_plus_df["acousticness"]


acousticness_all = df_2020_final_sorted_popularity["acousticness"]


acousticness_popular_p_value = ttest_ind(acousticness_popular, acousticness_all)

# Print rounded P Value

acousticness_popular_p_value = round(acousticness_popular_p_value[1], 4)

acousticness_popular_p_value = print(f' P Value: {acousticness_popular_p_value}')

acousticness_popular_p_value
```

```python
# T test for danceability

danceability_popular = seventy_plus_df["danceability"]


danceability_all = df_2020_final_sorted_popularity["danceability"]


danceability_popular_p_value = ttest_ind(danceability_popular, danceability_all)

# Print rounded P Value

danceability_popular_p_value = round(danceability_popular_p_value[1], 24)

danceability_popular_p_value = print(f' P Value: {danceability_popular_p_value}')

danceability_popular_p_value
```

```python
#Scatterplot of acousticness...

from scipy.stats import linregress

data_acousticness = df_2020_final_sorted_popularity["acousticness"]

data_popular_acousticness = df_2020_final_sorted_popularity["popularity"]

plt.scatter(data_acousticness, data_popular_acousticness, marker="o", facecolors="red", edgecolors="black", alpha=0.2)


# Add the linear regression equation and line to plot
(slope, intercept, rvalue, pvalue, stderr) = linregress(data_acousticness, data_popular_acousticness)
regress_values = data_acousticness * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(data_acousticness,regress_values)
plt.annotate(line_eq,(.3,50),fontsize=15,color="red")
print(f"The r-squared is: {rvalue}")
plt.xlabel('Acousticness Score')
plt.ylabel('Popularity')
plt.title("Relationship Between Acousticness & Popularity")
plt.savefig("Acousticness_Popularity_Relationship.png")
```
